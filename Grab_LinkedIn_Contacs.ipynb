```python
def loginLinkedin():
    
    global username, password

    driver.get('https://www.linkedin.com')
    time.sleep(2)
    driver.find_element_by_xpath('//div[@class="sign-in-form__inputs"]/.//input[@class="input__field input__field--with-label"][@name="session_key"]').click()
    driver.find_element_by_xpath('//div[@class="sign-in-form__inputs"]/.//input[@class="input__field input__field--with-label"][@name="session_key"]').send_keys(username)
    time.sleep(2)
    driver.find_element_by_xpath('//div[@class="sign-in-form__inputs"]/.//input[@class="input__field input__field--with-label"][@name="session_password"]').click()
    driver.find_element_by_xpath('//div[@class="sign-in-form__inputs"]/.//input[@class="input__field input__field--with-label"][@name="session_password"]').send_keys(password)
    time.sleep(2)
    driver.find_element_by_xpath('//form[@class="sign-in-form"]/.//button[@type="submit"][@class="sign-in-form__submit-btn"]').click()
    
    time.sleep(8)

    # test for incorrect credentials
    driver.implicitly_wait(5)
    pwd_error_list = driver.find_elements_by_xpath('//*[@id="error-for-password"]')
    driver.implicitly_wait(20)
    if len(pwd_error_list)>0:
        if driver.find_element_by_xpath('//*[@id="error-for-password"]').text != '':
            print('ALERT: incorrect password provided!')
            print(f'{username} has been recognized')
            password = getpass.getpass(f'please provide correct password for {username}')
            pwd_input = driver.find_element_by_xpath('//input[@id="password"]')
            pwd_input.click()
            pwd_input.clear()
            pwd_input.send_keys(password)  
            driver.find_element_by_xpath('//button[text()="Sign in"]').click()
            time.sleep(8)
        elif driver.find_element_by_xpath('//*[@id="error-for-username"]').text != '':
            print(f'ALERT: {username} NOT recognized!')
            print('please provide correct credentials')
            username = input('email or phone:')
            password = getpass.getpass('password:')
            name_input =driver.find_element_by_xpath('//input[@id="username"]')
            name_input.click()
            name_input.clear()
            name_input.send_keys(username)
            pwd_input = driver.find_element_by_xpath('//input[@id="password"]')
            pwd_input.click()
            pwd_input.clear()
            pwd_input.send_keys(password)
            driver.find_element_by_xpath('//button[text()="Sign in"]').click()
            time.sleep(8)

    # test for PIN challenge
    try:
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//span[@id="mynetwork-tab-icon"]')))
    except:
        pass
    driver.implicitly_wait(5)
    PIN_input_list = driver.find_elements_by_xpath('//input[@id="input__email_verification_pin"]')
    driver.implicitly_wait(20)

    if len(PIN_input_list) > 0:
        linkedin_PIN = input('EXTRA security precaution! A PIN has been sent to the email you provided, please provide the PIN too:')
        driver.find_element_by_xpath('//input[@id="input__email_verification_pin"]').send_keys(linkedin_PIN)
        driver.find_element_by_xpath('//button[@id="email-pin-submit-button"]').click()

    # test for good login
    # else 'f it, aks to retry
    try:
        WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//span[@id="mynetwork-tab-icon"]')))
    except:
        pass

    mynetwork_elems_count = len(driver.find_elements_by_xpath('//span[@id="mynetwork-tab-icon"]'))
    if mynetwork_elems_count == 1:
        print('GOOD login')
    elif mynetwork_elems_count == 0:
        print('BAD login')
        print("TRY THIS AGAIN PLEASE")
        
def grabConectionsListElements():
    RESULTS = list()

    driver.get("https://www.linkedin.com/search/results/people/?facetNetwork=[%22F%22]&origin=MEMBER_PROFILE_CANNED_SEARCH")

    #scroll down and up
    scheight = 1
    while scheight < 9.9:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight/%s);" % scheight)
        scheight += .1

    time.sleep(2)
    next_btn = driver.find_element_by_xpath('//button[contains(@class,"artdeco-pagination__button")][@aria-label="Next"]')
    time.sleep(2)
    search1sts = driver.find_elements_by_xpath('//div[@class="search-result__wrapper"]')

    pages_count = driver.find_element_by_xpath('//ul[@class="artdeco-pagination__pages artdeco-pagination__pages--number"]/li[last()]').text
    pages_count = int(pages_count)

    RESULTS += [card.get_attribute('outerHTML') for card in search1sts]

    for i in range(pages_count-1):
        next_btn = driver.find_element_by_xpath('//button[contains(@class,"artdeco-pagination__button")][@aria-label="Next"]')
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        next_btn.click()
        time.sleep(2)
        #scroll down and up
        scheight = 1
        while scheight < 9.9:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight/%s);" % scheight)
            scheight += .1
        time.sleep(2)
        next_btn = driver.find_element_by_xpath('//button[contains(@class,"artdeco-pagination__button")][@aria-label="Next"]')
        time.sleep(2)
        search1sts = driver.find_elements_by_xpath('//div[@class="search-result__wrapper"]')
        RESULTS += [card.get_attribute('outerHTML') for card in search1sts]
    
    # collect into a df
    df = pd.DataFrame()
    for res in RESULTS:
        resSoup = bs4.BeautifulSoup(res)
        name = resSoup.find('span',{'class':"name actor-name"}).text.strip()
        profile_href = resSoup.find('a',{'data-control-name':"search_srp_result"})['href']
        df = df.append({'name':name,'profile_href':profile_href},ignore_index=True)
    
    pickle.dump(df, open(CWD/f"{username}_contacts_df.pkl",'wb'))
    return df


def collectContactDetails(df):
    
    if 'contact_details' not in df.columns:
        df['contact_details'] = np.nan

    for idx,row in df.iterrows():
        if pd.isnull(df.loc[idx,'contact_details']):
            profile_href = row['profile_href']
            contact_details_href = f"https://www.linkedin.com{profile_href}detail/contact-info/"

            driver.get(contact_details_href)
            rand = round(random.uniform(3,6),1)
            time.sleep(rand)
            df.loc[idx,'contact_details'] = driver.find_element_by_xpath('//div[@class="pv-profile-section__section-info section-info"]').text

            pickle.dump(df, open(CWD/f"{username}_contacts_df.pkl",'wb'))
            
def parseContactDetails(df):

    tags = ['Connected',
            'Email',
            'Phone',
            'Birthday',
            'Twitter',
            'Website',
            'Websites',
            'IM',
            'Address']

    for idx,row in df.iterrows():
        elems = row['contact_details'].split('\n')
        idxs = {tag:elems.index(tag) for tag in tags if tag in elems}
        idxs = {k: v for k, v in sorted(idxs.items(), key=lambda item: item[1])}
        idxs['end'] = len(elems)
        keys = list(idxs.keys())

        #keys[keys.index(k)+1]
        res = {k: elems[v+1:idxs[keys[keys.index(k)+1]]] for k,v in idxs.items() if k != keys[-1]}
        for k,v in res.items():
            df.at[idx,k]='\n'.join(v)
```

```python
if not pathlib.Path(CWD/f'{username}_contacts_df.pkl').is_file():
    print('collecting a list of "connections" and their page urls')
    df = grabConectionsListElements()
    print('Collected a list of "connections" and their page urls')

collectContactDetails(df)
parseContactDetails(df)
df.to_excel(pathlib.Path(CWD/f'{username}_LinkedIn_contacts.xlsx'),index=False)
df.to_csv(pathlib.Path(CWD/f'{username}_LinkedIn_contacts.csv'),index=False)
```
